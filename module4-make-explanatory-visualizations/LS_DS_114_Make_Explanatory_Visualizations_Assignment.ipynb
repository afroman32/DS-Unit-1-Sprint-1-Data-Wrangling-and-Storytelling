```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fake_cable = pd.Series([3.1, 3.6, 2.8, 3.2, 2.5, 3.6, 4.1, 3.0, 3.5, 2.1, 1.1, 2.1, 1.5, 
                        1.8, 2.0, 1.8, 1.5, 1.8, 2.1, 1.6, 2.5, 4.6, 5.0, 1.5])
fake_online = pd.Series([1.1, 1.3, 1.3, 1.0, 1.1, 1.5, 1.3, 1.32, 1.6, 1.1, 1.0, 1.3, 
                         0.9, 0.9, 1.2, 0.9, 0.8, 1.1, 1.0, 1.0, 2.5, 4.6, 3.2, 2.6, 1.4,])
```

```python
# use fivethirtyeight style
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()


# plot the lines
fake_cable.plot(color = '#4EBAC3', linewidth = 3, )
fake_online.plot(color = '#DD7645', linewidth = 3)

# set title
ax.set_title(label = "Biden's Ukraine-related media bump is fading", size = 16, 
             fontweight = 'bold', x = .51, y = 1.2)

# Set subtitle
ax.text(s = "Biden's share of media mentions on each medium relative to the" +
        " next\nmost-mentioned candidate each week", x = -2, y = 5.5)

# set and format y tick labels
ax.set_yticklabels([' ',2,3,4,'5x'], fontsize = 12, color = 'gray')
ax.set_yticks([1,2,3,4,5])

# set and format x tick labels
ax.set_xticklabels([" ","MAY'19", 'JUNE', 'JULY', 'AUG.', 'SEPT.', 'OCT.', ' '], 
                   fontsize=12, color = 'gray')
ax.set_xticks(range(-1,24,4))

# cable news label
ax.text(s='Cable news', x = -.5, y = 3.9, fontweight = 'bold', fontsize = 12)

# online news label
ax.text(s='Online news', x = -1, y = 1.5, fontweight = 'bold', fontsize = 12)

# even label
ax.text(s = 'EVEN', x = -2, y = 1, fontsize = 10, color = 'gray')

# explanation label
ax.text(s = '           After the Ukraine news broke, \n       Biden was '+
        'mentioned            \nmore than the next most-mentioned \n            '
        +'       candidate on cable news', x = 8.5, y = 4.3, fontsize = 9)
ax.text(s = 'five times', x = 16.6, y = 4.72, fontsize = 8, fontweight = 'bold')

# add arrow
ax.arrow(19.5, 5, 1, 1, head_width=2, head_length = 2, shape = 'right')

# line for even
plt.plot([.1,25],[1,1], linewidth = 2, color = 'gray')

# Bottom stuff
ax.text(s = 'FiveThirtyEight', x = -2, y = -.5, fontsize = 8, color = 'gray')
ax.text(s = 'SOURCE: INTERNET TV NEWS ARCHIVE VIA GOELT API AND MEDIA CLOUD', 
        x = 11, y = -.5, fontsize = 8, color = 'gray')
plt.show()
```
